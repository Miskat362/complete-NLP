```python
corpus = "Hello Welcome,to my NLP practice file.Please check all file! to become expert in NLP."
print(corpus)
```

Sentence -> Paragraph [sent_tokenize]

```python
# Sentence --> paragraphs
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
documents = nltk.sent_tokenize(corpus)
print(documents)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

* Paragraph-->words [word_tokenize]
* sentence--->words [word_tokenize]

```python
# Paragraph --> words
# sentence ---> words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

This detect all punctuation [wordpunct_tokenize]

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

In this, full stop isn't detect except last sentence [TreebankWordTokenizer]

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
